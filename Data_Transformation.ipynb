```python
import pandas as pd
```

```python
df = pd.read_parquet('data/raw_data.parquet')
```

```python
df.info()
```

```python
df['year'] = df['release_date'].apply(lambda x: int(x.split('-')[0]))
```

```python
df.isnull().sum()
```

```python
df[['type_x','type_y']].value_counts()
```

```python
df[['uri_x','uri_y']].head()
```

```python
drop_cols = ['type_x','type_y','uri_y','duration_ms_y']
```

```python
df.drop(columns=drop_cols,inplace=True)
```

```python
df.rename(columns={'uri_x':'uri','duration_ms_x':'duration_ms'},inplace=True)
```

```python
df.head()
```

```python
df.select_dtypes(include=object).head()
```

```python
df['album_name'] = df['album'].apply(lambda x: x['name'])
df['artists_name'] = df['artists'].apply(lambda x: [i['name'] for i in x])
df['no_of_artists'] = df['artists'].apply(lambda x: len(x))
df['no_of_markets'] = df['available_markets'].apply(lambda x: len(x))
df['album_type'] = df['album'].apply(lambda x: x['album_type'])
df['album_total_tracks'] = df['album'].apply(lambda x: x['total_tracks'])
```

```python
df.head()
```

```python
cols_to_drop = ['album','artists','external_ids','external_urls','href','preview_url','uri','track_href','analysis_url']
```

```python
df.drop(columns=cols_to_drop,inplace=True)
```

```python
df.set_index('id',inplace=True)
df.head()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors = 'coerce')
```

```python
df.isnull().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.to_parquet('data/preprocessed_data.parquet')
```
