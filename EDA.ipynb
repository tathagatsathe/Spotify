```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams.update({'font.size': 10.0})
```

```python
df = pd.read_parquet('data/preprocessed_data.parquet')
df.head()
```

```python
df.groupby(['year','album_name']).count().sort_values(by='name').groupby(level=0).tail(1)
```

```python
popular_track_by_year = df.groupby([pd.Grouper(key='release_date', axis=0, freq='Y'),'album_name','name']).agg({'popularity':'max'}).sort_values(by='popularity').groupby(level=0).tail(1).reset_index(level=[1,2]).sort_index()
```

```python
popular_track_by_year.popularity.plot()
```

```python
artists_with_most_tracks_by_year = df.explode('artists_name').groupby([pd.Grouper(key='release_date', axis=0, freq='Y'),'artists_name'])[['name']].count().sort_values(by='name').groupby(level=0).tail(1).reset_index(level=1).sort_index().rename(columns={'name':'count'})
artists_with_most_tracks_by_year.index = artists_with_most_tracks_by_year.index.year
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(x=artists_with_most_tracks_by_year.index, y='count', data=artists_with_most_tracks_by_year,ax=ax)

for i, row in artists_with_most_tracks_by_year.iterrows():
    ax.text(i-1960, row['count']+2, str(row['artists_name']), ha='center', va='bottom', rotation=90, fontsize=8)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90,fontsize=10)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Tracks released')
ax.set_ylim(top=ax.get_ylim()[1] * 1.3)
plt.tight_layout()
plt.show()
```

```python
popular_artists_by_year = df.explode('artists_name').groupby([pd.Grouper(key='release_date', axis=0, freq='Y'),'artists_name']).agg({'popularity':'mean'}).sort_values(by='popularity').groupby(level=0).tail(1).reset_index(level=1).sort_index()
popular_artists_by_year.index = popular_artists_by_year.index.year
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(x=popular_artists_by_year.index, y='popularity', data=popular_artists_by_year,ax=ax)

for i, row in popular_artists_by_year.iterrows():
    ax.text(i-1960, row['popularity']+2, str(row['artists_name']), ha='center', va='bottom', rotation=90, fontsize=8)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90,fontsize=10)
ax.set_xlabel('Year')
ax.set_ylabel('Popularity')
ax.set_ylim(top=ax.get_ylim()[1] * 1.3)
plt.tight_layout()
plt.show()
```

```python
audio_feat_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                   'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
mean_by_yr = df.groupby(['year'])[[*audio_feat_cols,'popularity']].mean()
mean_by_yr
```

```python
fig, axes = plt.subplots(4,3,figsize=(20,20))
# fig.delaxes(axes[3][2])
fig.tight_layout(pad=3)

for i, ax in enumerate(axes.ravel()):
    sns.lineplot(data=mean_by_yr,x='year',y=mean_by_yr.columns[i],ax=ax).set(title=mean_by_yr.columns[i])
```

```python
fig, axes = plt.subplots(len(audio_feat_cols),figsize=(20,120))

for i, ax in enumerate(axes.ravel()):
    sns.boxplot(data = df , x='year',y=audio_feat_cols[i],ax=ax)
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90,fontsize=10)
    ax.set_xlabel('Year',fontsize=15)
    ax.set_ylabel(audio_feat_cols[i],fontsize=15)
```
